# Generar datos falsos en Python con Faker

```python
from faker import Faker
import pandas as pd
```

```python
fake = Faker(locale="es_ES")

def create_employees(num_employees):
    employee_list = []
    
    for _ in range(1, num_employees):
        employee = {}
        employee["first_name"] = fake.first_name()
        employee["last_name"] = fake.last_name()
        employee["job"] = fake.job()
        employee["department"] = fake.random_element(elements=("TI", "RRHH", "Marketing", "Finanzas"))
        employee["role"] = fake.random_element(elements=("Administrador", "Programador", "Analista", "Asociado"))
        employee["salary"] = fake.random_int(min=30000, max=150000, step=1000)
        employee["email"] = fake.email()
        employee_list.append(employee)
    
    return pd.DataFrame(employee_list)
```

```python
create_employees(20)
```

```python
df = create_employees(20)
```

```python
df.to_csv("../data/fake_employes.csv")
```
